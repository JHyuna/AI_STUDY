# 01. 결정 트리 (Decision Tree) 시각화

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')

# DecisionTree Classifier 생성
dt_clf = DecisionTreeClassifier(random_state=156)

# 붓꽃 데이터를 로딩하고, 학습과 테스트 데이터 셋으로 분리
iris_data = load_iris()
X_train , X_test , y_train , y_test = train_test_split(iris_data.data, iris_data.target,
                                                       test_size=0.2,  random_state=11)

# DecisionTreeClassifer 학습. 
dt_clf.fit(X_train , y_train)
```

- **export_graphviz()**: Graphviz가 읽어 들여서 그래프 형태로 시각화할 수 있는 출력 파일 생성
   - 파라미터: 학습 완료된 estimator, output 파일명, 결정 클래스 이름, 피처 이름

```python
from sklearn.tree import export_graphviz

# export_graphviz()의 호출 결과로 out_file로 지정된 tree.dot 파일을 생성함. 
export_graphviz(dt_clf, out_file="tree.dot", class_names=iris_data.target_names , \
feature_names = iris_data.feature_names, impurity=True, filled=True)
```

## 과적합 해결법 1. 최대 트리 깊이 제어 (max_depth)

```python
# max_depth가 3인 DecisionTree Classifier 생성하고 학습
dt_clf_max_depth = DecisionTreeClassifier(random_state=156, max_depth=3)
dt_clf_max_depth.fit(X_train , y_train)

# 시각화할 출력파일 생성하고 시각화
export_graphviz(dt_clf_max_depth, out_file="tree_max_depth.dot", class_names=iris_data.target_names , \
feature_names = iris_data.feature_names, impurity=True, filled=True)
with open("tree_max_depth.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

## 과적합 해결법 2. 자식 규칙 노드를 분할해 만들기 위한 최소한의 샘플 데이터 개수 제어 (min_sample_splits)

```python
# min_sample_split가 4인 DecisionTree Classifier 생성하고 학습
dt_clf_min_samples_split = DecisionTreeClassifier(random_state=156, min_samples_split=4)
dt_clf_min_samples_split.fit(X_train , y_train)

# 시각화할 출력파일 생성하고 시각화
export_graphviz(dt_clf_min_samples_split, out_file="tree_min_samples_split.dot", class_names=iris_data.target_names , \
feature_names = iris_data.feature_names, impurity=True, filled=True)
with open("tree_min_samples_split.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

#### ▶ 마지막 레벨의 2, 3번째 노드를 보면 아직 클래스 값이 완벽히 구별되지 않았음에도 샘플 개수가 3개밖에 없으므로 더 이상 분할하지 않는 걸 알 수 있음

## 과적합 해결법 3. 리프 노드가 될 수 있는 샘플 데이터 건수의 최솟값 지정 (min_samples_leaf)

```python
# min_sample_split가 4인 DecisionTree Classifier 생성하고 학습
dt_clf_min_samples_leaf = DecisionTreeClassifier(random_state=156, min_samples_leaf=4)
dt_clf_min_samples_leaf.fit(X_train , y_train)

# 시각화할 출력파일 생성하고 시각화
export_graphviz(dt_clf_min_samples_leaf, out_file="tree_min_samples_leaf.dot", class_names=iris_data.target_names , \
feature_names = iris_data.feature_names, impurity=True, filled=True)
with open("tree_min_samples_leaf.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

#### ▶ sample이 4 이상인 노드는 리프 클래스 노드가 될 수 있으므로 규칙이 sample 4인 노드를 만들 수 있는 상황을 반영하여 변경됨
#### ex) 레벨 3의 왼쪽 노드를 보면 이전 트리들과 규칙이 달라진 걸 볼 수 있는데 이는 샘플이 4 이상인 노드를 만들기 위함

> 결정 트리에선 **어떠한 속성을 규칙 조건으로 선택하느냐**가 중요함  
→ **feature_importances_ 속성**: 규칙을 정하는 데 있어 각 피처의 중요도를 제공

```python
import seaborn as sns
import numpy as np
%matplotlib inline

# feature importance 추출 
print("Feature importances:\n{0}".format(np.round(dt_clf.feature_importances_, 3)))

# feature별 importance 매핑
# 값 ∝ 중요도
for name, value in zip(iris_data.feature_names , dt_clf.feature_importances_):
    print('{0} : {1:.3f}'.format(name, value))

# feature importance를 column 별로 시각화 하기 
sns.barplot(x=dt_clf.feature_importances_ , y=iris_data.feature_names)
```

# 02. 결정 트리 과적합(Overfitting)

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
%matplotlib inline

plt.title("3 Class values with 2 Features Sample data creation")

# 2차원 시각화를 위해서 feature는 2개, 결정값 클래스는 3가지 유형의 classification 샘플 데이터 생성. 
# make_classification의 반환값 = 피처 데이터 세트, 클래스 레이블 데이터 세트
X_features, y_labels = make_classification(n_features=2, n_redundant=0, n_informative=2,
                             n_classes=3, n_clusters_per_class=1,random_state=0)

# plot 형태로 2개의 feature로 2차원 좌표 시각화, 각 클래스값은 다른 색깔로 표시됨. 
plt.scatter(X_features[:, 0], X_features[:, 1], marker='o', c=y_labels, s=25, cmap='rainbow', edgecolor='k')
```

```python
import numpy as np

# Classifier의 Decision Boundary를 시각화 하는 함수
def visualize_boundary(model, X, y):
    fig,ax = plt.subplots()
    
    # 학습 데이타 scatter plot으로 나타내기
    ax.scatter(X[:, 0], X[:, 1], c=y, s=25, cmap='rainbow', edgecolor='k',
               clim=(y.min(), y.max()), zorder=3)
    ax.axis('tight')
    ax.axis('off')
    xlim_start , xlim_end = ax.get_xlim()
    ylim_start , ylim_end = ax.get_ylim()
    
    # 호출 파라미터로 들어온 training 데이타로 model 학습 . 
    model.fit(X, y)
    # meshgrid 형태인 모든 좌표값으로 예측 수행. 
    xx, yy = np.meshgrid(np.linspace(xlim_start,xlim_end, num=200),np.linspace(ylim_start,ylim_end, num=200))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()]).reshape(xx.shape)
    
    # contourf() 를 이용하여 class boundary 를 visualization 수행. 
    n_classes = len(np.unique(y))
    contours = ax.contourf(xx, yy, Z, alpha=0.3,
                           levels=np.arange(n_classes + 1) - 0.5,
                           cmap='rainbow', clim=(y.min(), y.max()),
                           zorder=1)
```

```python
from sklearn.tree import DecisionTreeClassifier

# 특정한 트리 생성 제약없는 결정 트리의 Decsion Boundary 시각화.
dt_clf_default = DecisionTreeClassifier().fit(X_features, y_labels)
visualize_boundary(dt_clf_default, X_features, y_labels)
```

#### ▶ 아웃라이어들까지 제대로 잡아내기 위해 경계가 매우 많아짐 → 과적합 가능성 大

```python
# min_samples_leaf=6 으로 트리 생성 조건을 제약한 Decision Boundary 시각화
dt_clf_min_samples_leaf = DecisionTreeClassifier( min_samples_leaf=6).fit(X_features, y_labels)
visualize_boundary(dt_clf_min_samples_leaf, X_features, y_labels)
```

#### ▶ 아웃라이어들에 반응하지 않고 좀 더 일반화된 분류 규칙에 따라 분류됨
